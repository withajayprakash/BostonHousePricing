```python
# Dividing the datset into independent and dependent features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
y_train
```

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
## print the coefficients and the intercept
print(linear.coef_)
```

```python
print(linear.intercept_)
```

```python
## on which parameter the model has been trained
linear.get_params()
```

```python
## Prediction with the test data
lin_pred=linear.predict(x_test)
lin_pred
```

## Assumption

```python
## plot a scatter plot for the prediction 
plt.scatter(y_test,lin_pred)
```

```python
residuals=y_test-lin_pred
residuals
```

```python
## Plot this residuals
sns.displot(residuals,kind="kde")
```

```python
## Scatter plot with respect to residuals and prediction 
## Uniform Distribution 
plt.scatter(lin_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,lin_pred))
print(mean_squared_error(y_test,lin_pred))
print(np.sqrt(mean_squared_error(y_test,lin_pred)))
```

## R square and adjusted R square

```python
## R Square 
from sklearn.metrics import r2_score
score=r2_score(y_test,lin_pred)
print(score)
```

```python
## Adjusted R Square
1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
